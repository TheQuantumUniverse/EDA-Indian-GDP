### Data Preprocessing

* Since data for 2021-22, is not available for each column, we are neglecting that year in our study.

```python
df_2 = df_1.drop(labels = 17, axis = 0) # This will not alter the original dataframe
df_1
```

```python
df_2 # We can see that index [17] has been deleted in the new data frame
```

* We are also neglecting those columns, wherein data for all the years under consideration is not available.

```python
df_2.drop(labels = ['NVA at Basic Prices', 'GNP at Basic Prices', 'NNP at Basic Prices'], axis = 1, inplace = True)
```

```python
df_2.columns # We can see that the above mentioned columns have been deleted
```

* Now we will delete other columns which do not have data for all the years under consideration.

```python
df_2.drop(labels = ['Net National Disposable Income', 'per capita GNI at factor cost', 'Per Capita GNDI', 'Net Savings', 'Gross Savings', 'Gross National Disposable Income', 'Other current transfers (net) from ROW', 'rate of gross domestic saving*', 'Population(in million)'], axis = 1, inplace = True)
df_2
```

```python
df_2.info()
```

* We can see that many float values have been assigned 'object' data type. Let us change their data type to 'float'

```python
dates = ['Year']
data = list(df_2.drop('Year', axis = 1).columns)
for col in data:
    df_2[col]=pd.to_numeric(df_2[col], errors='coerce')
```

### Data Analysis

* Let us plot each of the columns against different years. Plot function from matplotlib will be useful for this.

```python
for col in df_2.drop('Year', axis = 1).columns:
    df_2.plot(x='Year', y = col, kind = 'line', figsize = (12,11))
    plt.show
```

* From theory, we know that GVA at Basic Prices, GDP at Market Prices, NDP at Market Prices, GNP at Market Prices and NNP at Market Prices should move together. Let us plot them together to see if it is true for the data.

```python
df_2.plot(x = 'Year', y = ['GVA at Basic Prices', 'GDP at Market Prices', 'NDP at Market Prices', 'GNP at Market Prices', 'NNP at Market Prices'], kind = 'line', figsize = (12,11))
```

* We see that the graphs almost move together. We also see that the absolute difference between GNP and GDP or Net Factor Income from Abroad (NFIA) has been increasing over the years. Let us now plot a heatmap of correlation co-efficients between different macroeconomic aggregates.

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_2.drop('Year', axis=1).corr(), cmap='coolwarm',annot=True)
plt.show()
```

* We see how most of the values are perfectly correlated with each other. This is backed up by theory. Let us now create another dataframe which will calculate percentage change over the years in each of the variable.

```python
df_3 = df_2.drop('Year', axis = 1).pct_change().mul(100)
df_3
```

```python
first_column = df_2['Year']
first_column
```

```python
df_3.insert(0, 'Year', first_column) # This will insert 'Year' as the first column in the new dataframe. 
```

```python
df_3
```

* Let us plot growth in GDP at Market Prices

```python
df_3.plot('Year', 'GDP at Market Prices', kind = 'line', figsize = (12,10))
```

* We are able to see a sharp decline in growth in 2008-09 due to the Global Financial Crisis and in 2020-21 due to the after effects of Global Coronavirus Pandemic. This completes our exploratory analysis of data on different macro economic aggregates.
